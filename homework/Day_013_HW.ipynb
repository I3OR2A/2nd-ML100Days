```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# 設定 data_path
dir_data = './data/'
```

### 之前做過的處理

```python
# 讀取資料檔
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
app_train.shape
```

```python
# 將只有兩種值的類別型欄位, 做 Label Encoder, 計算相關係數時讓這些欄位可以被包含在內
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# 檢查每一個 column
for col in app_train:
    if app_train[col].dtype == 'object':
        # 如果只有兩種值的類別型欄位
        if len(list(app_train[col].unique())) <= 2:
            # 就做 Label Encoder, 以加入相關係數檢查
            app_train[col] = le.fit_transform(app_train[col])            
print(app_train.shape)
app_train.head()
```

```python
# 受雇日數為異常值的資料, 另外設一個欄位記錄, 並將異常的日數轉成空值 (np.nan)
app_train['DAYS_EMPLOYED_ANOM'] = app_train["DAYS_EMPLOYED"] == 365243
app_train['DAYS_EMPLOYED'].replace({365243: np.nan}, inplace = True)

# 出生日數 (DAYS_BIRTH) 取絕對值 
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
```

## 練習時間

```python
import seaborn as sns
```

```python
print('---Before---')
print(app_train['DAYS_BIRTH'].head())
app_train['DAYS_BIRTH_YEAR'] = app_train['DAYS_BIRTH'] / 365
print('---After---')
print(app_train['DAYS_BIRTH_YEAR'])
```

```python
app_train['TARGET']
```

```python
app_train['DAYS_BIRTH_YEAR_GRP'] = pd.cut(app_train['DAYS_BIRTH_YEAR'], 10)
# 計算每個年齡區間的 Target、DAYS_BIRTH與 YEARS_BIRTH 的平均值
birth_groups  = app_train.groupby('DAYS_BIRTH_YEAR_GRP').mean()
birth_groups
```

```python
# 以年齡區間為 x, target 為 y 繪製 barplot
px = app_train['DAYS_BIRTH_YEAR_GRP']
py = app_train['TARGET']
sns.barplot(px, py) 

# Plot labeling
plt.xticks(rotation = 45); 
plt.xlabel('Birth Group (years)'); 
plt.ylabel('Failure to Repay (%)')
plt.title('Failure to Repay by Birth Group');
```
